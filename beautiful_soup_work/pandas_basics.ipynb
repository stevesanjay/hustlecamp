```python
average_score=df[["math_score",	"history_score","physics_score","chemistry_score","biology_score","english_score","geography_score"]].mean()
print("Average of math_score	history_score	physics_score	chemistry_score	biology_score	english_score	geography_score")
print(average_score)
```

```python
highest_scores_subjectwise = df[['math_score', 'history_score', 'physics_score', 'chemistry_score', 'biology_score', 'english_score', 'geography_score']].max()


print("Highest scores achieved in each subject:")
print(highest_scores_subjectwise)
```

```python
lowest_scores_subjectwise = df[['math_score', 'history_score', 'physics_score', 'chemistry_score', 'biology_score', 'english_score', 'geography_score']].min()
print("Lowest scores achieved in each subject:")
print(lowest_scores_subjectwise)
```

```python
# Calculate the lowest score across all subjects
lowest_score = df[['math_score', 'history_score', 'physics_score', 'chemistry_score', 'biology_score', 'english_score', 'geography_score']].min()

# Print the lowest score
print(f"The lowest score is {lowest_score}")
```

```python
gender_counts = df['gender'].value_counts()
print("Male students:", gender_counts)
```

```python
print("Female students:", gender_counts)
```

```python
average_score_male = df[df['gender'] == 'male'][['math_score', 'history_score', 'physics_score', 'chemistry_score', 'biology_score', 'english_score', 'geography_score']].mean()
print(f"The average score for male students is {average_score_male}")
```

```python
# Filter the DataFrame to include only students who scored above 90 in any subject
scores_above_90 = df[(df['math_score'] > 90) |
                     (df['history_score'] > 90) |
                     (df['physics_score'] > 90) |
                     (df['chemistry_score'] > 90) |
                     (df['biology_score'] > 90) |
                     (df['english_score'] > 90) |
                     (df['geography_score'] > 90)]

# Print the scores of students who scored above 90 in any subject
print(scores_above_90)
```

```python
# Count the number of students who passed (scored above 50) in any subject
passed_students_count = len(df[(df['math_score'] > 50) |
                               (df['history_score'] > 50) |
                               (df['physics_score'] > 50) |
                               (df['chemistry_score'] > 50) |
                               (df['biology_score'] > 50) |
                               (df['english_score'] > 50) |
                               (df['geography_score'] > 50)])

# Calculate the total number of students
total_students = len(df)

# Calculate the percentage of students who passed
percentage_passed = (passed_students_count / total_students) * 100

# Print the percentage of students who passed
print("Percentage of students who passed:", percentage_passed)
```
